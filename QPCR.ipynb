```python
# Example usage in Jupyter notebook
from qpcr_analyzer import QPCRAnalyzer

# Initialize analyzer
analyzer = QPCRAnalyzer(
    file_path='sample_qpcr_data.csv',  # or .xlsx
    target_genes=['Gene1', 'Gene2','Gene3'],  # your target genes
    reference_gene='GAPDH',  # your reference gene
    #exclude_groups=['']  # optional: groups to exclude
)

# Get statistics
stats = analyzer.get_statistics()
display(stats)

# Create visualizations
analyzer.plot_results()

# Save all results to Excel
analyzer.save_results()
```
